### Assessing Data

**Data Duplikat**

Pertama-tama, dalam penilaian kualitas data akan melihat jumlah data yang merupakan duplikat.

```python
# Memberi label pada program yang akan digunakan
total_data = len(cust_items)
data_duplikat = cust_items.duplicated().sum()
usable_data = total_data - data_duplikat
```

```python
# Untuk melihat jumlah data yang merupakan data duplikat dan membandingkannya dengan total data
print(f"Jumlah data: {total_data}")
print(f"Jumlah data duplikat: {data_duplikat}")
print(f"Data yang dapat digunakan: {usable_data}")
```

***Missing Value***

Tahap berikutnya dalam *assessing data* adalah melihat jumlah data yang kosong.

```python
# Untuk melihat jumlah data yang tidak memiliki nilai dan membandingkannya dengan total data
usable_data_notnull = cust_items.notnull().sum()
print(f"Jumlah data yang tidak memiliki nilai: {usable_data_notnull}")
```

**Outlier Check**

Pada bagian ini, bagian terakhir, akan melihat penyebaran titik data yang berada sangat jauh dari titik data lain.

```python
q25, q75 = np.percentile(cust_items['price'], [25, 75])
iqr = q75 - q25
lower_bound = q25 - (iqr * 1.5)
upper_bound = q75 + (iqr * 1.5)

outliers = cust_items[(cust_items['price'] < lower_bound) | (cust_items['price'] > upper_bound)]
print(f"Jumlah data yang memiliki outlier: {len(outliers)}")
```

### Cleaning Data

**Pembersihan data duplikat**

Pertama-tama dalam proses pembersihan data, hal pertama yang dilakukan adalah membersihkan data dari data-data duplikat. Alasannya karena data yang digunakan memiliki banyak sekali duplikat, sehingga dapat memberi data yang tidak valid

```python
cust_items.drop_duplicates(inplace=True)
```

**Pembersihan data kosong**

Langkah berikutnya adalah membersihkan data-data yang kosong. Agar meningkatkan validitas data dan keberhasilan menjawab pertanyaan, data-data dengan nilai yang kosong harus dibersihkan agar tidak memberi hasil akhir yang buruk/tidak dapat menjawab pertanyaan.

```python
cust_items.dropna(axis=0, inplace=True)
```

**Pembersihan data *outliers***

Tahap terakhir dalam pembersihan data kali ini adalah membersihkan data dari data-data yang *outliers*. Hal ini penting agar penyebaran data yang digunakan menjadi lebih baik.

```python
kondisi_lower_than = cust_items['price'] < lower_bound
kondisi_more_than = cust_items['price'] > upper_bound
cust_items.drop(cust_items[kondisi_lower_than].index, inplace=True)
cust_items.drop(cust_items[kondisi_more_than].index, inplace=True)
```

```python
# Kamus yang berisi penjelasan singkatan dari masing-masing wilayah
state_descriptions = {
    'AC': 'Acre',
    'AL': 'Alagoas',
    'AP': 'Amapá',
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Ceará',
    'DF': 'Distrito Federal',
    'ES': 'Espírito Santo',
    'GO': 'Goiás',
    'MA': 'Maranhão',
    'MT': 'Mato Grosso',
    'MS': 'Mato Grosso do Sul',
    'MG': 'Minas Gerais',
    'PA': 'Pará',
    'PB': 'Paraíba',
    'PR': 'Paraná',
    'PE': 'Pernambuco',
    'PI': 'Piauí',
    'RJ': 'Rio de Janeiro',
    'RN': 'Rio Grande do Norte',
    'RS': 'Rio Grande do Sul',
    'RO': 'Rondônia',
    'RR': 'Roraima',
    'SC': 'Santa Catarina',
    'SP': 'São Paulo',
    'SE': 'Sergipe',
    'TO': 'Tocantins'
}

# Menghitung value counts dari kolom 'geolocation_state'
customer_state_counts = cust_items['customer_state'].value_counts()

# Membuat bar plot
plt.figure(figsize=(12, 6))  # Menyesuaikan ukuran figure jika diperlukan
plt.bar(x=customer_state_counts.index, height=customer_state_counts.values)

# Mengatur rotasi pada label sumbu x
plt.xticks(ticks=range(len(customer_state_counts.index)),
           labels=[state_descriptions.get(state, state) for state in customer_state_counts.index],
           rotation=90)

# Menambahkan judul dan label sumbu
plt.title('Penyebaran pembeli berdasarkan provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah pembeli')

# Menampilkan plot
plt.tight_layout()  # Menyesuaikan layout agar tidak terpotong
plt.show()
```

```python
# Menghitung value counts dari kolom 'geolocation_state'
customer_city_counts = cust_items['customer_city'].value_counts().nlargest(10)

# Membuat bar plot
plt.figure(figsize=(12, 6))  # Menyesuaikan ukuran figure jika diperlukan
plt.barh(y=customer_city_counts.index[::-1], width=customer_city_counts.values[::-1])

# Menambahkan judul dan label sumbu
plt.title('Penyebaran pembeli berdasarkan kota')
plt.xlabel('Jumlah Pembeli')
plt.ylabel('Kota')

# Menampilkan plot
plt.tight_layout()  # Menyesuaikan layout agar tidak terpotong
plt.show()
```

### Pertanyaan 2: Produk seperti apa yang marak dibeli pada masing-masing daerah?

Untuk mendapat *insight* dari visualisasi yang akan dilakukan agar dapat menjawab pertanyaan tersebut, terdapat beberapa visualisasi yang harus dilakukan, yaitu

- Produk yang paling banyak dibeli
- Penjualan produk berdasarkan provinsi
- Penjualan produk berdasarkan kota

```python
# Menghitung value counts dari kolom 'geolocation_state'
product_counts = cust_items['product_category_name_english'].value_counts().nlargest(10)

# Palet warna yang vibrant dan warm
colors = ['#9b59b6', '#3498db', '#95a5a6', '#e74c3c', '#34495e', '#2ecc71', '#f39c12', '#16a085', '#d35400', '#c0392b']

# Membuat bar plot
plt.figure(figsize=(12, 6))  # Menyesuaikan ukuran figure jika diperlukan
plt.pie(product_counts.values, labels=product_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85, colors=colors)

# Menambahkan judul dan label sumbu
plt.title('Penyebaran pembeli berdasarkan kota', pad=20)
plt.axis('equal')

# Menampilkan plot
plt.tight_layout()  # Menyesuaikan layout agar tidak terpotong
plt.show()
```

```python
# Menghitung jumlah produk terbanyak yang dibeli berdasarkan provinsi
top_product_by_state = cust_items.groupby(['customer_state', 'product_category_name_english']).size().reset_index(name='count')
top_product_by_state = top_product_by_state.sort_values(by=['customer_state', 'count'], ascending=[True, False])

# Menghitung total pembelian per provinsi untuk mendapatkan 5 provinsi teratas
total_purchases_per_state = top_product_by_state.groupby('customer_state')['count'].sum().nlargest(5).index

# Memfilter data untuk hanya menyertakan 5 provinsi teratas
top_product_by_state_top5 = top_product_by_state[top_product_by_state['customer_state'].isin(total_purchases_per_state)]

#
top_product_by_state_top5 = top_product_by_state_top5.groupby('customer_state').head(3)

# Ganti singkatan provinsi dengan nama lengkap menggunakan dictionary
top_product_by_state_top5['customer_state'] = top_product_by_state_top5['customer_state'].map(state_descriptions)

# Membuat plot menggunakan seaborn
plt.figure(figsize=(12, 8))
sns.barplot(data=top_product_by_state_top5, x='customer_state', y='count', hue='product_category_name_english')
plt.title('Produk Terbanyak yang Dibeli Berdasarkan Provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.legend(title='Kategori Produk', bbox_to_anchor=(1.05, 1), loc='upper left')

# Membalikkan urutan sumbu x agar provinsi diurutkan dari yang paling sedikit ke yang paling banyak
plt.gca().invert_xaxis()

plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah produk terbanyak yang dibeli berdasarkan provinsi
top_product_by_city = cust_items.groupby(['customer_city', 'product_category_name_english']).size().reset_index(name='count')
top_product_by_city = top_product_by_city.sort_values(by=['customer_city', 'count'], ascending=[True, False])

# Menghitung total pembelian per provinsi untuk mendapatkan 5 provinsi teratas
total_purchases_per_city = top_product_by_city.groupby('customer_city')['count'].sum().nlargest(5).index

# Memfilter data untuk hanya menyertakan 5 provinsi teratas
top_product_by_city_top5 = top_product_by_city[top_product_by_city['customer_city'].isin(total_purchases_per_city)]

#
top_product_by_city_top5 = top_product_by_city_top5.groupby('customer_city').head(3)


# Membuat plot menggunakan seaborn
plt.figure(figsize=(12, 8))
sns.barplot(data=top_product_by_city_top5, x='customer_city', y='count', hue='product_category_name_english')
plt.title('Produk Terbanyak yang Dibeli Berdasarkan Kota')
plt.xlabel('Kota')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.legend(title='Kategori Produk', bbox_to_anchor=(1.05, 1), loc='upper left')

# Membalikkan urutan sumbu x agar provinsi diurutkan dari yang paling sedikit ke yang paling banyak
plt.gca().invert_xaxis()

plt.tight_layout()
plt.show()
```
